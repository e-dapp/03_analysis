The same way that we started working with post number 2 (graphing), let's start this one off by quickly showing the fruits of our labor for the previous post. Remember, we're working with the data on deaths collected in 122 US cities, which can be found at https://catalog.data.gov/dataset/table-iii-deaths-in-122-u-s-cities. 

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

deaths = pd.read_csv('./data/deaths.csv')
deaths_no_flag = deaths.drop([col for col in deaths.columns if 'flag' in col], axis = 1)

areas = {}
unique_areas = pd.unique(deaths_no_flag['Reporting Area'].ravel())
for area in unique_areas:
    areas[area] = deaths_no_flag[deaths_no_flag['Reporting Area'] == area]

for area in unique_areas:
    areas[area] = areas[area].set_index(areas[area]['MMWR WEEK'])
for area in unique_areas:
    areas[area] = areas[area].drop([col for col in areas[area].columns if 'MMWR' in col], axis = 1)
    
weeks = {}
unique_weeks = pd.unique(deaths_no_flag['MMWR WEEK'])
for week in unique_weeks:
    weeks[week] = deaths_no_flag[deaths_no_flag['MMWR WEEK'] == week]

for week in unique_weeks:
    weeks[week] = weeks[week][weeks[week]['Reporting Area'] != 'Total\xc2\xb6']

super_regions = ['New England', 'Mid. Atlantic', 'E.N. Central', 
                 'W.N. Central', 'S. Atlantic', 'E.S. Central', 'W.S. Central',
                 'Mountain', 'Pacific']

for week in unique_weeks:
    weeks[week] = weeks[week].drop([row for row in weeks[week].index 
                    if weeks[week]['Reporting Area'].ix[row] in super_regions], axis = 0)
    
weeks_sorted = {}
for week in unique_weeks:
    weeks_sorted[week] = weeks[week].sort('All causes, by age (years), All Ages', ascending = False)

def quick_sorted_bar(n, e):
    '''takes n = int, e = int
        plots top e cities for week n'''
    
    plt.figure(figsize=(15, 10))
    fig = sns.barplot(x = weeks_sorted[n][:e]['Reporting Area'], 
                      y = weeks_sorted[n][:e]['All causes, by age (years), All Ages'])
    fig.set(xlabel = 'Reporting Area', ylabel = 'Deaths - All Ages')

for week in unique_weeks:
    weeks_sorted[week] = weeks_sorted[week][pd.notnull(weeks_sorted[week]['All causes, by age (years), All Ages'])]

quick_sorted_bar(1, 10)
```

This is more or less what we accomplished last time, and in more or less the order we did it. In this post we're going to take a little look at the P&I (Pneumonia and Influenza) column, and do a bit of analysis on that sucker. Specifically we'll be working in the areas dictionary. This way we will eventually be able to see what percentage of all deaths are caused by Pnuemonia or Influenza throughout the year for any specific area in our data. We'll culminate by looking at our results for the Total of all areas. Let's see the dataframe for Boston, since it's near the top of the list.

```python
areas['Boston, MA'][:5]
```

You see, for each week we have the deaths by age group and we also have the information for how many of these deaths were due to Pneumonia or Influenza. Note that the P&I column corresponds to the total deaths.

We're going to be using bar graphs a lot in this post, so let's try to make a 'fit's-all' function to expedite graphing and labelling.

```python
def bar_graph(x, y, x_lab, y_lab):
    '''takes pd.Series: x, y and str: x_lab, y_lab
        and plots beautiful beautiful graphs.'''
    
    plt.figure(figsize=(15, 10))
    fig = sns.barplot(x = x, 
                      y = y)
    fig.set(xlabel = x_lab, ylabel = y_lab)
```

Not sure that this will actually make things any easier, but we're just function writing machines right now. As always, let's test it.

```python
bar_graph(areas['Boston, MA'].index, areas['Boston, MA']['P&I† Total'], 'Week in Year','P&I Total')
```

Word, everything seems to be in order. But how's about we make a new set of dataframes that focus on the information we want.

```python
areas_pi = {}
for area in unique_areas:
    areas_pi[area] = areas[area][['Reporting Area', 'All causes, by age (years), All Ages', 'P&I† Total']]
```

What I'm doing here is going through and grabbing the columns pertaining to the area, the total deaths, and the number of those deaths that were caused by Pneumonia or Influenza. Since we're so interested in expediting the little things, let's rename the columns of these dataframes.

```python
for area in unique_areas:
    areas_pi[area].columns = ['area', 'deaths', 'p&i']
```

Sanity check:

```python
areas_pi['Boston, MA'][:5]
```

Certainly looks like we're in business. Do note that we could do this for the larger dataframes if we were so inclined. In fact, we may well deem it necessary when we want to extract information from some of the other columns. We are interested, for the moment, in what role P&I deaths play in the grand scheme of all the deaths. So let's do a little calculation and give it its own column.

```python
for area in unique_areas:
    areas_pi[area]['%'] = 100.0 * areas_pi[area]['p&i'] / areas_pi[area]['deaths']
```

```python
areas_pi['Boston, MA'][:5]
```

What we've just added is the percent of the total deaths each week that were caused by Pneumonia or Influenza. It's worthwhile to note that this information is now accessible to us over all surveyed Reporting Areas. That means that aside from having it for 'Boston, MA', 'New York, NY' and other cities, we also have this information for all of what we've dubbed Super Regions. These should help us take a look at the whole country in broad strokes. First let's augment our bar_graph function to make use of the hue parameter to act like a graph title, so we can distinguish the graphs from one another.

```python
def bar_graph2(x, y, x_lab, y_lab, h = None):
    '''takes pd.Series: x, y; str: x_lab, y_lab
        and pd.Series h for hue catagories
        which we're using as labels'''
    
    plt.figure(figsize=(15, 10))
    fig = sns.barplot(x = x, 
                      y = y,
                      hue = h)
    fig.set(xlabel = x_lab, ylabel = y_lab)
```

Now let's look at the graphs for the percentage P&I deaths make up per week, for each of the Super Regions.

```python
for region in super_regions:
    bar_graph2(areas_pi[region].index, areas_pi[region]['%'], 
               'Week in Year', 'P&I Death Percentage', 
               areas_pi[region]['area'])
```

In general, the winter months seem to have a higher percentage of the total deaths being made up of those caused by Pneumonia or Influenza. One last visualization for this post: Let's take all the reporting areas as a whole and look at the Total entry in the areas_pi dictionary. But let's be clean about it. And by that, I mean, let's get rid of the wiggity whack characters that we ended up with because whoever entered this data wanted to make sure we knew that '¶ Total includes unknown ages.' While I appreciate that information, there was probably a better way of doing it than throwing a '¶' at us. In any case, it's nothing to get too bothered about. Let's inspect our dictionary keys.

```python
areas_pi.keys()
```

The key we're interested in is 'Total' followed by some characters which actually represent the symbol '¶'. That string of characters is very easy for a computer read and write, but happily (and sometimes unhappily) we are not computers. Humans won't in general be able to reproduce that string on command, because it means nothing to us. Let's replace that entry in the dictionary with a properly titled one.

```python
areas_pi['Total'] = areas_pi.pop('Total\xc2\xb6')
```

Easy peasy. But we're still going to want to replace the instances of 'Total¶' in the 'Total' dataframe itself.

```python
areas_pi['Total'] = areas_pi['Total'].replace('Total\xc2\xb6', 'Total')
areas_pi['Total'][:5]
```

We seem to be right on track. The reason I'm going through the trouble of replacing all the 'Total\xc2\xb6's is because one of seaborn or matplotlib is of a similar mind to me when it comes to this special character bollox. GRAPH TIME!

```python
bar_graph2(areas_pi['Total'].index, 
          areas_pi['Total']['%'], 
          'Week in Year', '% P&I',
          areas_pi['Total']['area'])
```

Here we can clearly see the trend of P&I deaths that, in the aggregate, account for higher percentages of all deaths in the winter months. Most people would probably guess that this is the case, but now we've got data to support that claim. Word.
